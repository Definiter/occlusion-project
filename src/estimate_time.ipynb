```python
import os
import time
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (10, 10)

with open('/home/haow3/log.txt') as f:
    lines = f.readlines()

x = []
y = []
    
start_time = None
start_sum = None
for line in lines:
    words = line.split(' ')
    h = int(words[1].split(':')[0])
    m = int(words[1].split(':')[1])
    s = int(words[1].split(':')[2][0:2])
    #if h <= 19:
    #    h += 24
    now_time = h * 3600 + m * 60 + s
    
    if start_time == None:
        start_time = now_time
    if start_sum == None:
        start_sum = int(words[-2])
    
    x.append(now_time - start_time)
    y.append(int(words[-2]))
    
    last_time = now_time - start_time
    last_sum = int(words[-2])

def print_time(t):
    print int(t / 3600 / 24), time.strftime("%H:%M:%S", time.gmtime(t))


print_time(last_time)
    
all_time = float(last_time) / last_sum * 343360
print_time(all_time)

rest_time = all_time - last_time
print_time(rest_time)

ax = plt.gca()
ax.set_ylabel('number of images')
ax.set_xlabel('time (s)')
plt.plot(x, y)
```
